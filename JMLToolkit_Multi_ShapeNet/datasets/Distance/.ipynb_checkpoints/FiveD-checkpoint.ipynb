```python
def getTimeseriesLabel():
    # Using readlines() 
    file1 = open('./../ALT_AND_AFP_ARFF/ALT_AND_AFP_TRAIN.arff', 'r') 
    file2 = open('./../ALT_AND_AFP_ARFF/ALT_AND_AFP_TEST.arff', 'r') 
    LinesArr = []
    Lines1 = file1.readlines()
    Lines2 = file2.readlines()
    LinesArr.append(Lines1)
    LinesArr.append(Lines2)

    labelArr = []
    # Strips the newline character 
    for Lines in LinesArr:
        count = 0
        startFlag = False
        for line in Lines: 
        #     print("Line{}: {}".format(count, line.strip()))
            line = line.strip()
            if startFlag:
                line = line.split("\\n")
                for subStr in line:
                    subSubStr = subStr.split(",")
                    if len(subSubStr) == 16:
                        # Get the label of each multi-timeseries
                        label = subSubStr[15]
        #                 print(label)
                        labelArr.append(float(label))

        #         print('count: ' + str(count))
        #         print('--------')
                count += 1

            if(line=='@data'):
                startFlag = True

    print(len(labelArr))
    return labelArr
```

```python
df1 = pd.read_csv('./feature_train.txt', sep=" ", header=None)
df1.columns = ['A', 'B', 'C', 'D', 'E']
df1
```

```python
df2 = pd.read_csv('./feature_test.txt', sep=" ", header=None)
df2.columns = ['A', 'B', 'C', 'D', 'E']
# df2
frame = [df1, df2]
df = pd.concat(frame)
df['label'] = getTimeseriesLabel()
df
```

```python
from sklearn.linear_model import Perceptron

X = df.drop(['label'], axis=1).values
y = np.ones(df.shape[0])
y[df['label']!=0.0] = -1

# Split X and y into training and testing sets 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

ppn = Perceptron(max_iter=100)

# Train the perceptron
ppn.fit(X_train, y_train)

y_pred = ppn.predict(X_test)

print("test accuracy: " + str(sum(y_test == y_pred)/len(y_pred)))
```

```python
def outputParam(fname, coef):
    file1 = open(fname,"w")
    # \n is placed to indicate EOL (End of Line) 
    coefList = coef.flatten().tolist()
    for i in range(len(coefList)):
        aCoef = coefList[i]
        file1.write(str(aCoef))
        if i < len(coefList) - 1:
            file1.write(', ')
    file1.close() #to change file access modes
```

```python
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
clf = SVC(C = 1e5, kernel = 'linear')
clf.fit(X_train, y_train)
w = clf.coef_
b = clf.intercept_
print('w = ',w)
print('b = ',b)
print('Indices of support vectors = ', clf.support_)
print('Support vectors = ', clf.support_vectors_)
print('Number of support vectors for each class = ', clf.n_support_)
print('Coefficients of the support vector in the decision function = ', np.abs(clf.dual_coef_))
```

```python
clf.score(X_test, y_test)
```

```python
# .sort()
arr = []
for i in range(df.shape[0]):
    powerSum = 0
    alist = df.iloc[i:i+1,0:5].values.flatten().tolist()
    alist.sort()
    for j in range(5):
        powerSum += alist[j]**2
#     print(powerSum.values[0])
    arr.append(powerSum)
        
df['powerSum'] = arr
# df
df
```
