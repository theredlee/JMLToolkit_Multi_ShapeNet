```python
def clearMissingVal_1(df):
    '''
    df.isna().any(axis=1)
    '''
    return df[(df['SubjectID']!='#NULL!') & (df['HCC']!='#NULL!') & (df['Sex']!='#NULL!') & (df['LISReferenceDatetime']!='#NULL!') & (df['Age']!='#NULL!') & (df['AFP_LISResultNumericResult_aug']!='#NULL!') & (df['ALT_LISResultNumericResult']!='#NULL!') & (df['Alb_LISResultNumericResult']!='#NULL!') & (df['Tb_LISResultNumericResult']!='#NULL!') & (df['Plt_LISResultNumericResult']!='#NULL!')]

clear_df_1 = clearMissingVal_1(df)
clear_df_1
```

```python
def clearMissingVal_2(my_df):
    df_2 = my_df.groupby(by=['SubjectID']).count()
    return my_df[my_df['SubjectID'].isin(df_2[df_2['HCC'] >= 15].reset_index().SubjectID.values)]

clear_df_2 = clearMissingVal_2(clear_df_1)
clear_df_2
```

```python
def testing4(clear_df_2):
    clear_df_2['HCC'] = clear_df_2['HCC'].astype(int)
    HCC = clear_df_2.groupby(by=['SubjectID']).mean()
    return HCC[HCC['HCC']==1]
testing4(clear_df_2)
```

```python
def pointProcessing(my_df, size):
    my_df[['HCC','AFP_LISResultNumericResult_aug','ALT_LISResultNumericResult','Alb_LISResultNumericResult', 'Tb_LISResultNumericResult', 'Plt_LISResultNumericResult']] = my_df[['HCC', 'AFP_LISResultNumericResult_aug','ALT_LISResultNumericResult','Alb_LISResultNumericResult', 'Tb_LISResultNumericResult', 'Plt_LISResultNumericResult']].astype(np.float32)
    my_user_list = []
    users = list(set(my_df.SubjectID.values))
    for usr in users:
        user_df = my_df[my_df['SubjectID']==usr]
        my_list = []
        # HCC = 
        my_list.append(user_df.HCC.values.tolist())
        # aft = 
        my_list.append(user_df.AFP_LISResultNumericResult_aug.values.tolist())
        # alt = 
        my_list.append(user_df.ALT_LISResultNumericResult.values.tolist())
        # alb = 
        my_list.append(user_df.Alb_LISResultNumericResult.values.tolist())
        # tb = 
        my_list.append(user_df.Tb_LISResultNumericResult.values.tolist())
        # plt = 
        my_list.append(user_df.Plt_LISResultNumericResult.values.tolist())
        
        # Initialize counter_index
        counter_index = len(my_list[0])-1
        while(True):
            # The lengths of five arrays in arr are identical
            if len(my_list[0]) > size and counter_index-1>=0:
                for i in range(len(my_list)):
                    dim = my_list[i]
                    last_index = counter_index
                    last_sencond_index = last_index-1
                    # Pop the last two elements
                    a = dim.pop(last_index)
                    b = dim.pop(last_sencond_index)
                    avg = statistics.mean([a, b])
                    # Assign the mean val into last_sencond_index position
                    dim.insert(last_sencond_index, avg)
                # Update counter by minus 2
                counter_index -= 2
            elif counter_index<0:
                # After traveling the whole list, update the counter for a new loop
                counter_index = len(my_list[0])-1
                print("Updated counter_index: " + str(counter_index) + ", updated length: " + str(len(my_list[0])))
            else:
                break
        
        # Ater processing each user's data, add them into a new array        
        my_user_list.append(my_list)
        
    return my_user_list

def testing1(size=15):
    processed_5Dim_data = pointProcessing(clear_df_2, size)
    pd.DataFrame(processed_5Dim_data[0])

def testing2(my_df,size=15):
    return pointProcessing(my_df, size)
#     pd.DataFrame(processed_5Dim_data[0])
```

```python
def smaple_equal_size(my_df):
    my_df['HCC'] = my_df['HCC'].astype(int)
    HCC = my_df.groupby(by=['SubjectID']).mean()
    sample_size = min(HCC[HCC['HCC']==0].shape[0], HCC[HCC['HCC']==1].shape[0])
    a = HCC[HCC['HCC']==0]
    b = HCC[HCC['HCC']==1]
    user_a = a.sample(frac=sample_size/a.shape[0]).reset_index().SubjectID.values.tolist()
    user_b = b.sample(frac=sample_size/b.shape[0]).reset_index().SubjectID.values.tolist()
    print("Number of user_a: " + str(len(user_a)))
    print("Number of user_b: " + str(len(user_b)))
    df_a = my_df[my_df.SubjectID.isin(user_a)]
    df_b = my_df[my_df.SubjectID.isin(user_b)]
    return pd.concat([df_a, df_b])
    
def list_splitter(list_to_split, ratio):
    elements = len(list_to_split)
    middle = int(elements * ratio)
    return [list_to_split[:middle], list_to_split[middle:]]

def train_test_split(my_df, ratio):
    user = my_df.groupby(by=['SubjectID']).count().reset_index().SubjectID.values.tolist()
    user_train, user_test = list_splitter(user, ratio)
    print("user_train: " + str(len(user_train)))
    print("user_test: " + str(len(user_test)))
    return my_df.loc[my_df.SubjectID.isin(user_train)], my_df.loc[my_df.SubjectID.isin(user_test)]
        
def testing3():
    df_test = smaple_equal_size(clear_df_2)
    # df_test = smaple_equal_size(clear_df_2).groupby(by=['SubjectID']).count()
    # df_test[df_test.HCC == 1]
    df_test
    
def testing_formal_1(my_df):
    df_updated = smaple_equal_size(my_df)
    df_train , df_test = train_test_split(df_updated, 0.5)
    
    print("df_updated num of patients: " + str(df_updated.groupby(by=['SubjectID']).mean().shape[0]))
    print("df_train num of patients: " + str(df_train.groupby(by=['SubjectID']).mean().shape[0]))
    print("df_test num of patients: " + str(df_test.groupby(by=['SubjectID']).mean().shape[0]))
    
    size=15
    # df_test = smaple_equal_size(clear_df_2).groupby(by=['SubjectID']).count()
    # df_test[df_test.HCC == 1]
    return (testing2(df_train, size), testing2(df_test, size))
```

```python
output_data_train, output_data_test = testing_formal_1(clear_df_2)
# output_data
```

```python
def output_1(processed_5Dim_data, name):
    f = open(name + "_dim_.txt", "w+")
    for i in range(len(processed_5Dim_data)):
        one_user_data = processed_5Dim_data[i]
        f.write("'")
#         label = '{:.18e}'.format(-1)
        label = -1
        for k in range(len(one_user_data)):
            item = one_user_data[k]
            if k == 0:
                label = '{:.18e}'.format(item[0])
#                 label = item[0]
            else:
                for j in range(len(item)):
                    val = '{:.18e}'.format(item[j])
#                     val = item[j]
                    if j < len(item)-1:
                        f.write("%s," % val)
                    else:
                        f.write("%s" % val)
                if k < len(one_user_data)-1:
                    f.write("\\n")
                else:
                    f.write("'")
        f.write(",%s\n" % label)
    f.close()

# output_1(output_data_train, "train")
# output_1(output_data_test, "test")
    
def output_2(processed_5Dim_data, name):
    multidim = []
    labels = []
    dim_size = 5
    for i in range(dim_size):
        arr = []
        multidim.append(arr)
    '''
    dim_1 = []
    dim_2 = []
    dim_3 = []
    dim_4 = []
    dim_5 = []
    '''
    for i in range(len(processed_5Dim_data)):
        dim_n = []
        one_user_data = processed_5Dim_data[i]
        for k in range(len(one_user_data)):
            one_dim = one_user_data[k]
            if k == 0:
                label = one_dim[0]
                labels.append(label)
            else:
                # k-1 since the index of dim 0-5 are from 1-6
                multidim[k-1].append(one_dim)
    
    for i in range(len(multidim)):
        f = open(name + "_dim_" + str(i) + ".txt", "w+")
        one_dim_all_users = multidim[i]
        label = '{:.18e}'.format(-1)
        for k in range(len(one_dim_all_users)):
            label = '{:.18e}'.format(labels[k])
            # Write label at first
            f.write("%s," % label)
            
            # Then write vals
            one_dim = one_dim_all_users[k]
            for j in range(len(one_dim)):
                val = '{:.18e}'.format(one_dim[j])
                if j < len(one_dim)-1:
                    f.write("%s," % val)
                else:
                    f.write("%s\n" % val)
        f.close()
    
print("size_train: " + str(len(output_data_train)))
print("size_test: " + str(len(output_data_test)))
output_1(output_data_train, "train")
output_1(output_data_test, "test")
```
