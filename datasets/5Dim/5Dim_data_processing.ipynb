```python
def clearMissingVal_1(df):
    '''
    df.isna().any(axis=1)
    '''
    return df[(df['SubjectID']!='#NULL!') & (df['HCC']!='#NULL!') & (df['Sex']!='#NULL!') & (df['LISReferenceDatetime']!='#NULL!') & (df['Age']!='#NULL!') & (df['AFP_LISResultNumericResult_aug']!='#NULL!') & (df['ALT_LISResultNumericResult']!='#NULL!') & (df['Alb_LISResultNumericResult']!='#NULL!') & (df['Tb_LISResultNumericResult']!='#NULL!') & (df['Plt_LISResultNumericResult']!='#NULL!')]

clear_df_1 = clearMissingVal_1(df)
clear_df_1
```

```python
describe_1 = clear_df_1.groupby(by=['SubjectID']).count().describe()
describe_1
```

```python
def clearMissingVal_2(my_df):
    df_2 = my_df.groupby(by=['SubjectID']).count()
    return my_df[my_df['SubjectID'].isin(df_2[df_2['HCC'] >= 15].reset_index().SubjectID.values)]

clear_df_2 = clearMissingVal_2(clear_df_1)
clear_df_2
```

```python
def pointProcessing(my_df, size):
    my_df[['AFP_LISResultNumericResult_aug','ALT_LISResultNumericResult','Alb_LISResultNumericResult', 'Tb_LISResultNumericResult', 'Plt_LISResultNumericResult']] = my_df[['AFP_LISResultNumericResult_aug','ALT_LISResultNumericResult','Alb_LISResultNumericResult', 'Tb_LISResultNumericResult', 'Plt_LISResultNumericResult']].astype(np.float32)
    my_user_list = []
    users = list(set(my_df.SubjectID.values))
    for usr in users:
        user_df = my_df[my_df['SubjectID']==usr]
        my_list = []
        # aft = 
        my_list.append(user_df.AFP_LISResultNumericResult_aug.values.tolist())
        # alt = 
        my_list.append(user_df.ALT_LISResultNumericResult.values.tolist())
        # alb = 
        my_list.append(user_df.Alb_LISResultNumericResult.values.tolist())
        # tb = 
        my_list.append(user_df.Tb_LISResultNumericResult.values.tolist())
        # plt = 
        my_list.append(user_df.Plt_LISResultNumericResult.values.tolist())
        
        # Initialize counter_index
        counter_index = len(my_list[0])-1
        while(True):
            # The lengths of five arrays in arr are identical
            if len(my_list[0]) > size and counter_index-1>=0:
                for i in range(len(my_list)):
                    dim = my_list[i]
                    last_index = counter_index
                    last_sencond_index = last_index-1
                    # Pop the last two elements
                    a = dim.pop(last_index)
                    b = dim.pop(last_sencond_index)
                    avg = statistics.mean([a, b])
                    # Assign the mean val into last_sencond_index position
                    dim.insert(last_sencond_index, avg)
                # Update counter by minus 2
                counter_index -= 2
            elif counter_index<0:
                # After traveling the whole list, update the counter for a new loop
                counter_index = len(my_list[0])-1
                print("Updated counter_index' length: " + str(counter_index))
            else:
                break
        
        # Ater processing each user's data, add them into a new array        
        my_user_list.append(my_list)
        
    return my_user_list

size = 15
processed_5Dim_data = pointProcessing(clear_df_2, size)
```
